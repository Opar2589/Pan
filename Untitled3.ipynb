```python
from gplearn.genetic import SymbolicRegressor
import numpy as np
import matplotlib.pyplot as plt

# Define and train the symbolic regressor
est_gp = SymbolicRegressor(
    population_size=1000,
    generations=20,
    stopping_criteria=0.01,
    function_set=['add', 'sub', 'mul', 'div', 'sqrt', 'log', 'sin', 'cos'],
    parsimony_coefficient=0.01,
    verbose=1,
    random_state=42
)
# Fit the model on the data
est_gp.fit(X_scaled, y_scaled)

# Define complexity and performance metrics
def evaluate_equation(program, X, y_true):
    try:
        y_pred = program.execute(X)
        mse = np.mean((y_pred - y_true) ** 2)  # Mean Squared Error
        complexity = len(str(program))  # Length of the equation
        return mse, complexity
    except:
        # Handle cases where the program fails to execute (e.g., division by zero)
        return float('inf'), float('inf')

# Evaluate Pareto front
pareto_front = []
for program in est_gp._programs[-1]:  # Access the last generation's programs
    if program is not None:  # Exclude uninitialized individuals
        mse, complexity = evaluate_equation(program, X_scaled, y_scaled)
        pareto_front.append((program, mse, complexity))

# Select non-dominated solutions
non_dominated = []
for i, (prog1, mse1, comp1) in enumerate(pareto_front):
    dominated = False
    for j, (prog2, mse2, comp2) in enumerate(pareto_front):
        if mse2 <= mse1 and comp2 <= comp1 and (mse2 < mse1 or comp2 < comp1):
            dominated = True
            break
    if not dominated:
        non_dominated.append((prog1, mse1, comp1))

# Display Pareto optimal solutions
print("\nPareto Optimal Equations:")
for program, mse, complexity in non_dominated:
    print(f"Equation: {program}, MSE: {mse:.4f}, Complexity: {complexity}")

# Visualize the Pareto front
mse_vals = [mse for _, mse, _ in non_dominated]
comp_vals = [comp for _, _, comp in non_dominated]

plt.scatter(mse_vals, comp_vals, label="Pareto Optimal")
plt.xlabel("Mean Squared Error")
plt.ylabel("Complexity")
plt.title("Pareto Front")
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Assuming inputs (X) and target (y) are defined
inputs = data[['V', 'd', 'υ', 'C', 'ρ', 'Hv', 'σy', 'KIC', 'E']]
target = data['Re']

# Normalize the data
scaler_X = StandardScaler()
scaler_y = StandardScaler()
X_scaled = scaler_X.fit_transform(inputs)
y_scaled = scaler_y.fit_transform(target.values.reshape(-1, 1)).flatten()

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)

# Train the model
model = LinearRegression()  # Replace with your actual regression model
model.fit(X_train, y_train)

# Predict on the test set
y_pred = model.predict(X_test)

# Denormalize predictions and actual values
y_test_denorm = scaler_y.inverse_transform(y_test.reshape(-1, 1)).flatten()
y_pred_denorm = scaler_y.inverse_transform(y_pred.reshape(-1, 1)).flatten()

# Calculate metrics
r2 = r2_score(y_test_denorm, y_pred_denorm)
rmse = np.sqrt(mean_squared_error(y_test_denorm, y_pred_denorm))

# Print metrics
print(f"R²: {r2:.4f}")
print(f"RMSE: {rmse:.4f}")

# Parity plot (Actual vs Predicted)
plt.figure(figsize=(8, 6))
plt.scatter(y_test_denorm, y_pred_denorm, alpha=0.7, label="Predicted vs Actual")
plt.plot([min(y_test_denorm), max(y_test_denorm)], [min(y_test_denorm), max(y_test_denorm)], 'r--', label="Ideal Fit")
plt.xlabel("Actual Re")
plt.ylabel("Predicted Re")
plt.title("Parity Plot")
plt.legend()
plt.grid()
plt.show()

# Residual plot (Error vs Predicted)
residuals = y_test_denorm - y_pred_denorm
plt.figure(figsize=(8, 6))
plt.scatter(y_pred_denorm, residuals, alpha=0.7)
plt.axhline(0, color='r', linestyle='--')
plt.xlabel("Predicted Re")
plt.ylabel("Residuals (Actual - Predicted)")
plt.title("Residuals Plot")
plt.grid()
plt.show()
```
